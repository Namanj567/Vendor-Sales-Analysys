```python
Purchase_summary = pd.read_sql_query("""
SELECT 
    pr.VendorNumber,
    pr.VendorName,
    pr.Brand,
    pr.PurchasePrice,
    pp.Price AS Actual_sellingPrice,
    pp.Volume,
    pp.Description,
    SUM(pr.Quantity) AS total_purchase_quantity,
    SUM(pr.Dollars) AS total_purchases_amount_in_dollars
FROM purchases AS pr 
JOIN purchase_prices AS pp
    ON pr.Brand = pp.Brand 
WHERE pr.PurchasePrice > 0
GROUP BY pr.VendorNumber, pr.VendorName, pr.Brand, pr.PurchasePrice, pp.Price, pp.Volume, pp.Description
ORDER BY total_purchases_amount_in_dollars ;
""", conn)

print(Purchase_summary)
```

```python
import time 
start_time = time.time()
vendor_sales_summary1 = pd.read_sql_query("""
WITH freight_cost_summary AS (
    SELECT VendorNumber, SUM(Freight) AS freight_Cost
    FROM vendor_invoice
    GROUP BY VendorNumber
),
Purchase_summary AS (
    SELECT 
        pr.VendorNumber,
        pr.VendorName,
        pr.Brand,
        pr.PurchasePrice,
        pp.Price AS Actual_sellingPrice,
        pp.Volume,
        pp.Description,
        SUM(pr.Quantity) AS total_purchase_quantity,
        SUM(pr.Dollars) AS total_purchases_amount_in_dollars
    FROM purchases AS pr 
    JOIN purchase_prices AS pp
        ON pr.Brand = pp.Brand 
    WHERE pr.PurchasePrice > 0
    GROUP BY pr.VendorNumber, pr.VendorName, pr.Brand, pr.PurchasePrice, pp.Price, pp.Volume
    ORDER BY total_purchases_amount_in_dollars
),
Sales_summary AS (
    SELECT 
        VendorNo, 
        Brand, 
        SUM(SalesPrice) AS TOTAL_SALES_PRICE, 
        SUM(SalesDollars) AS TOTAL_SALES_DOLLARS,
        SUM(ExciseTax) AS TOTALSALES_EXCISE_TAX, 
        SUM(SalesQuantity) AS TOTAL_SALES_QUANTITY
    FROM sales 
    GROUP BY VendorNo, Brand
)
SELECT 
    ps.VendorNumber, 
    ps.VendorName, 
    ps.Brand, 
    ps.PurchasePrice, 
    ps.Actual_sellingPrice, 
    ps.Volume,
    ps.Description,
    ps.total_purchase_quantity, 
    ps.total_purchases_amount_in_dollars,
    ss.TOTAL_SALES_QUANTITY,
    ss.TOTAL_SALES_PRICE, 
    ss.TOTAL_SALES_DOLLARS, 
    ss.TOTALSALES_EXCISE_TAX,
    fcs.freight_Cost
FROM Purchase_summary AS ps 
LEFT JOIN Sales_summary AS ss 
    ON ps.VendorNumber = ss.VendorNo
    AND ps.Brand = ss.Brand 
LEFT JOIN freight_cost_summary AS fcs 
    ON ps.VendorNumber = fcs.VendorNumber 
ORDER BY ps.total_purchases_amount_in_dollars;
""", conn)


end_time = time.time() 
vendor_sales_summary1
# '''this is the most important work of a data analyst to extract the data and then usko transform karke required 
# columns ko ek jagah load karna '''
```

```python
'''this query generates vendor wise sales and purchase summmary , which is valueable for :
 
 PERFORMANCE OPTIMIZATION
 -> the query involves heavy joins and aggregations on large dataset like sales and purchase .
 -> storing the pre-aggregated result avoid repetive expensive computation 
 ->helps i analyze sales ,purchase and price for different vendor and brands .
 -> future benefits of storing this data for faster dashboarding and reporting 
 -> instead of running hard query each time , dashboard can fetch data quickly from vendor sales summary .
 '''
```

```python
vendor_sales_summary1['Volume'] = vendor_sales_summary1['Volume'].astype('float64')
```

```python
vendor_sales_summary1.fillna(0,inplace=True)
```

```python
vendor_sales_summary1['VendorName'] = vendor_sales_summary1['VendorName'].str.strip()
```

```python
# creating some new columns 
# grossprofit 

vendor_sales_summary1['GrossProfit'] = vendor_sales_summary1['TOTAL_SALES_DOLLARS'] - vendor_sales_summary1['total_purchases_amount_in_dollars']
```

```python
vendor_sales_summary1['Stockturnover'] = vendor_sales_summary1['TOTAL_SALES_QUANTITY']/vendor_sales_summary1['total_purchase_quantity']
```

```python
vendor_sales_summary1['ProfitMargin'] =( vendor_sales_summary1['GrossProfit'] / vendor_sales_summary1['TOTAL_SALES_DOLLARS']) * 100
```

```python
vendor_sales_summary1['SalestoPurchaseRatio'] = vendor_sales_summary1['TOTAL_SALES_DOLLARS'] / vendor_sales_summary1['total_purchases_amount_in_dollars']
```
